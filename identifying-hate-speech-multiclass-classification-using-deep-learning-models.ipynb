```python
import re

def safe_replace(comment, replacements):
    pattern = re.compile(r'\b(' + '|'.join(re.escape(key) for key in replacements.keys()) + r')\b')
    replaced_words = []
    def replacer(match):
        word = match.group()
        replaced_words.append((word, replacements[word]))
        return replacements[word]
    replaced_comment = pattern.sub(replacer, comment)
    if replaced_words:
        print(f"Replacements in '{comment}': {replaced_words}")
    return replaced_comment

combined_df['Comment'] = combined_df['Comment'].apply(lambda x: safe_replace(x, replacements))
```

```python
from imblearn.over_sampling import RandomOverSampler
import pandas as pd

X_train = df['Comment']
y_train = df['Type']

classes = ["Appearance", "Not Hate", "Others", "Racial", "Religious", "Sexual", "Slang"]
class_values = [0, 1, 2, 3, 4, 5, 6]

target_count = 2000

balanced_df = pd.DataFrame()

for class_name, class_value in zip(classes, class_values):
    minority_mask = (y_train == class_value)

    # Reshape X_train for RandomOverSampler
    X_train_reshaped = X_train.values.reshape(-1, 1)

    random_over_sampler = RandomOverSampler(sampling_strategy='auto')

    X_resampled, y_resampled = random_over_sampler.fit_resample(X_train_reshaped, y_train)

    combined_df_resampled = pd.DataFrame({'content': X_resampled.flatten(), 'label': y_resampled})

    class_indices = combined_df_resampled[combined_df_resampled['label'] == class_value].index[:target_count]
    class_balanced_df = combined_df_resampled.loc[class_indices]

    balanced_df = pd.concat([balanced_df, class_balanced_df], ignore_index=True)

balanced_df = balanced_df.sample(frac=1).reset_index(drop=True)

print(balanced_df['label'].value_counts())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Embedding, GRU, Dense, Bidirectional, LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
texts = balanced_df['content'].values
labels = balanced_df['label'].values
```

```python
from tensorflow.keras.utils import to_categorical


labels= to_categorical(labels)
```

```python
len(texts)
```

```python
max_words = 6229
max_sequence_length = 300
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
data = pad_sequences(sequences, maxlen=max_sequence_length)
```

```python

#labels = np.asarray(labels)


train_texts, test_texts, train_labels, test_labels = train_test_split(data, labels, test_size=0.2, random_state=42)
```

#LSTM

```python
model = Sequential()
model.add(Embedding(max_words, 300, input_length=max_sequence_length))
model.add(LSTM(64))
model.add(Dense(7, activation='softmax'))


model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])


print(model.summary())
```

```python
history = model.fit(train_texts, train_labels, epochs=100, batch_size=64, validation_split=0.2)

loss, accuracy = model.evaluate(test_texts, test_labels)
print("Loss:", loss)
print("Accuracy:", accuracy)
```
