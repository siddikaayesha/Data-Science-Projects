### **Metod 1**

```python
# Importing the dataset
dataset = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
import seaborn as sns
sns.heatmap(dataset.isnull())
```
